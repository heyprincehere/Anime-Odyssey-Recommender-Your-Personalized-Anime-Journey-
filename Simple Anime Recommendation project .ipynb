Now we will import the dataset

```python
data=pd.read_csv('anime.csv')
```

```python
data=data[data['genre']!='isekai']
```

Let's create the algorithm to recommeding anime on the basisof genre

to know all the genre of the anime will use the command

```python
genre=data['genre'].values
```

```python
data.info
```

```python
lis=[]
for i in genre:
    i=str(i)
    for p in i.split(','):
        if p not in lis:
            lis.append(p)
```

```python
len(lis)
```

```python
dic={}
```

```python
t=[]
for i in genre:
    for j in lis:
        dic[j]=0
    i=str(i)
    for p in i.split(','):
        dic[p]+=1
    t.append(list(dic.values()))
```

```python
len(t[0])
```

```python
# input 1 is the anime name
# input 2 is the number of search result you want
s=input('Enter the anime you like and we will find more like those for you  :')
num=int(input('Please enter the number of recommendations you want:'))
name=data['name'].values
for i in range(len(name)):
    name[i]=name[i].lower()
s=s.lower()
h=-1
for i in range(len(name)):
    if s in name[i]:
        h=i
        break
imp=[]
if h==-1:
    print('Sorry No match found :(')
else:
    for i in range(len(t)):
        if i==h:
            continue
        else:
            if len(imp)<num:
                imp.append([distance.euclidean(t[i], t[h]),t[i],i])
            else:
                imp.sort()
                if imp[num-1][0]>distance.euclidean(t[i],t[h]):
                    del imp[num-1]
                    imp.append([distance.euclidean(t[i],t[h]),t[i],i])
    name=data['name'].values
    print('The anime recommended for you are :')
    count=0
    for i in imp:
        count+=1
        print(count,'. ',name[i[2]])
```
